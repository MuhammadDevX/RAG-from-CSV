```python
# Install LangChain Unstructured (which requires unstructured under the hood)

import os
from unstructured.partition.pdf import partition_pdf
import glob
import pandas as pd
```

```python
files = glob.glob(pathname='./**/*.pdf',recursive=True) # I have set recursive = True so that we can check subdirectories too.
print(len(files)) # confirm that you have all of the pdfs here with the correct path
```

```python
cases = [""] * 991
```

```python
len(cases)
```

```python
cases[0]
```

```python

for file_path in files:
  case_index = file_path.split("case-")[1].split(".pdf")[0]
  print(case_index)
```

```python
for file_path in files:
  case_index = int(file_path.split("case-")[1].split(".pdf")[0])
  elements = partition_pdf(file_path,strategy="fast")
  content = ""
  for element in elements:
    if element.category == "Title":
      content+=f"{element.text}\n"
    else:
      content+=element.text
  print("case number",case_index,"Completed")
  cases[case_index] = content
  
```

```python
remaining_indexes_for_ocr = []
```

```python
for i in range(len(cases)):
  if len(cases[i]) > 0:
    continue
  else:
    remaining_indexes_for_ocr.append(i)
```

```python
remaining_file_paths= []
```

```python
for file_path in files:
  if int(file_path.split("case-")[1].split(".pdf")[0]) in remaining_indexes_for_ocr:
    remaining_file_paths.append(file_path)
```

```python
len(remaining_file_paths)
```

```python
for file_path in remaining_file_paths:
  case_index = int(file_path.split("case-")[1].split(".pdf")[0])
  content = ""
  elements = partition_pdf(file_path,strategy="ocr_only")
  for element in elements:
    if element.category == "Title":
      content+=f"{element.text}\n"
    else:
      content+=element.text
  print("case number",case_index,"Completed")
  cases[case_index] = content
    
```

```python
dict = {"cases":cases}
```

```python
import pandas as pd
df = pd.DataFrame(dict)
```

```python
df.to_csv("cases.csv",index=False,header=True)
```
